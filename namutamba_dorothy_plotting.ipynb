```python
import matplotlib.pyplot as plt

# Line plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
data = {'x': x, 'y': y}
sns.scatterplot(data=data, x='x', y='y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
import plotly.graph_objects as go
from plotly.offline import plot

# Scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
fig = go.Figure(data=go.Scatter(x=x, y=y, mode='markers'))
fig.update_layout(title='Scatter Plot', xaxis_title='X-axis', yaxis_title='Y-axis')
plot(fig, filename='scatter_plot.html')

# Bar plot
x = ['A', 'B', 'C', 'D']
y = [10, 20, 15, 25]
fig = go.Figure(data=go.Bar(x=x, y=y))
fig.update_layout(title='Bar Plot', xaxis_title='X-axis', yaxis_title='Y-axis')
plot(fig, filename='bar_plot.html')
```

```python
import pandas as pd
df2 = pd.read_csv('UceResultsBySchool2011-2016.csv')
print(df2)
```

```python
shape = df2.shape
print("Shape of the dataset: ", shape)
```

```python
columns = df2.columns
print("Columns in the dataset: ", columns)
```

```python

# Remove white spaces and replace with underscore in column names
df2.columns = df2.columns.str.replace(' ', '_')

# Display the updated column names
print(df2.columns)
```

```python

# Get the column names
columns = df2.columns

# List to store duplicate rows
duplicate_rows = []

# Iterate over each column
for column in columns:
    # Get the unique values in the column
    unique_values = df2[column].unique()
    
    # Iterate over each value in the column
    for value in unique_values:
        # Check if the value is duplicated
        if df2[df2[column] == value].shape[0] > 1:
            # Append the duplicated rows to the list
            duplicate_rows.append(df2[df2[column] == value])

# Concatenate the duplicate rows into a single DataFrame
if duplicate_rows:
    duplicates_df = pd.concat(duplicate_rows)
    print("Duplicate rows:")
    print(duplicates_df)
else:
    print("No duplicates found.")
```

```python
null_counts = df2.isnull().sum(axis=1)
num_rows_with_null = null_counts[null_counts > 0].count()
print("Number of rows with null values:", num_rows_with_null)

```

```python
df2.dropna(thresh=None, subset=['SCHOOL'], inplace=True)
dropped_rows = df2[df2['SCHOOL'].isnull()]
print("Dropped Rows:")
print(dropped_rows)
```
